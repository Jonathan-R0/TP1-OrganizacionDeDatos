```python
df_info = df.describe().T
df_info
```

```python
df_info['range'] = df_info[['min', 'max']].apply(tuple, axis=1)
df_info.drop(columns=['count', 'std', 'min', '25%', '50%', '75%', 'max'], inplace = True)
df_info
```
